# 使用python将josn转换为shp文件并保存

```python
import geopandas
import matplotlib.pyplot as plt

data = geopandas.read_file('/mnt/f/shape_data/china.json')  #geopandas读取的文件可以直接plot
data.plot()
plt.show()
"""保存为shp"""
data.to_file('/mnt/f/shape_data/china_shapefile', driver='ESRI Shapefile', encoding='utf-8')
```
